```python
# All of the plotting function excpect np.array or np.ma.masked_array
# pandas or np.matrix best to convert these to np.array

# 1. u can use OO-style to, Use this in non-interactive case
# 2. u can rely on pyplot automatically create and manage the figures and axes, use this in interactive case
x = np.linspace(0, 2, 100)
fig, ax = plt.subplots()
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Simple Plot')
ax.legend()
```
